```python
# Set plot style
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 12))

# Age Distribution
sns.histplot(df['Age'], kde=True, ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Age Distribution')

# Academic Pressure Distribution
sns.histplot(df['Academic Pressure'], kde=True, ax=axes[0, 1], color='green')
axes[0, 1].set_title('Academic Pressure Distribution')

# Study Satisfaction Distribution
sns.histplot(df['Study Satisfaction'], kde=True, ax=axes[1, 0], color='orange')
axes[1, 0].set_title('Study Satisfaction Distribution')

# Study Hours Distribution
sns.histplot(df['Study Hours'], kde=True, ax=axes[1, 1], color='red')
axes[1, 1].set_title('Study Hours Distribution')

# Financial Stress Distribution
sns.histplot(df['Financial Stress'], kde=True, ax=axes[2, 0], color='purple')
axes[2, 0].set_title('Financial Stress Distribution')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Depression prevalence
depression_counts = df['Depression'].value_counts()

plt.figure(figsize=(6, 4))
sns.barplot(x=depression_counts.index, y=depression_counts.values, palette='rocket')
plt.title('Depression Prevalence')
plt.xlabel('Depression (Yes/No)')
plt.ylabel('Count')
plt.show()
```

```python
#Gender distribution on gender
# Depression by gender
sns.countplot(x='Depression', hue='Gender', data=df, palette='viridis')
plt.title('Depression Status by Gender')
plt.xlabel('Depression')
plt.ylabel('Count')
plt.show()
```

Strongest correlation is between Academic Pressure and Financial Stress (positive correlation).
Study Satisfaction negatively correlates with Academic Pressure and Financial Stress.
Other correlations are relatively weak.

```python
# Set plot style
sns.set(style="whitegrid")

# List of categorical features to visualize
categorical_features = ['Have you ever had suicidal thoughts ?', 'Family History of Mental Illness']
target_feature = 'Depression'

# Create subplots for each categorical feature
plt.figure(figsize=(12, len(categorical_features) * 5))

for i, feature in enumerate(categorical_features):
    plt.subplot(len(categorical_features), 1, i + 1)
    sns.countplot(x=feature, hue=target_feature, data=df, palette="coolwarm")
    plt.title(f"Comparison of {feature} with {target_feature}")
    plt.xlabel(feature)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Retain only numerical features
numerical_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Feature Correlation Matrix')
plt.show()
```
