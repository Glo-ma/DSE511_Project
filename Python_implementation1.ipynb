```python
# Replace 'Yes' and 'No' with 1 and 0
binary_columns = ["Have you ever had suicidal thoughts ?", "Family History of Mental Illness", "Depression"]
df[binary_columns] = df[binary_columns].replace({"Yes": 1, "No": 0})

# Replace 'Healthy', 'Unhealthy', and 'Moderate' with 0, 1, and 2
df["Dietary Habits"] = df["Dietary Habits"].replace({"Healthy": 0, "Unhealthy": 1, "Moderate": 2})

encode = LabelEncoder()
df_categorical =[]
for column in df.columns:
    if df[column].dtypes=="object":
        df_categorical.append(column)
for column in df_categorical:
    df[column] = encode.fit_transform(df[column])
```

```python
# Display the standardized dataset
df
```

```python
print(df[['Age', 'Study Hours', 'Academic Pressure', 'Financial Stress']].describe())
```

```python
df.to_csv('DSS_preprocessed.csv', index=False)

print("One-hot encoded dataset exported successfully!")
```
